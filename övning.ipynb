```python
def räkneord():
    ord= input(str(""))
    j = 0
    choice = input(str(""))
    choice = choice.lower()
    ord = ord.lower()
    ord = ord.split()
    for i in ord:
        if choice in i:
            j+=1


    print(f" Antal Förekomster av '{choice}': {j}")

räkneord()
```
