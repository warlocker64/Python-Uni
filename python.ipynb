```python
def kvinna_man(data):
    print("År och länder när kvinnorna presterar bättre än männen")
    print("--------------------------------------------------------")
    print("År\tLand\tKvinnor\tMän")
    print("--------------------------------------------------------")
    
    for i in range(1, len(data[0])):
        year = data[0][i]
        for j in range(2, len(data)):
            country = data[j][0]
            women_score = data[j][i] if data[j][1] == 'F' else None
            men_score = data[j][i] if data[j][1] == 'M' else None
                
            if women_score is not None and men_score is not None and women_score > men_score:
                print("%d\t%s\t%d\t%d" % (year, country, women_score, men_score))

# Exempeldata
data = [
    ['', 2018, 2018, 2015, 2015, 2012, 2012, 2009, 2009, 2006, 2006, 2003, 2003],
    ['', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'],
    ['Australia', 488, 494, 491, 497, 498, 510, 509, 519, 513, 527, 522, 527],
    ['Austria', 492, 505, 483, 510, 494, 517, 486, 506, 494, 517, 502, 509],
    ['Belgium', 502, 514, 500, 514, 509, 520, 504, 526, 517, 524, 525, 533],
    ['Brazil', 379, 388, 370, 385, 380, 397, 379, 394, 361, 380, 348, 365]
]

# Anropa funktionen med exemplet på data
kvinna_man(data)
```
