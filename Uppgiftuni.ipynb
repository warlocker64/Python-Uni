```python
# Skriv en inledande kommentar som talar om vad programmet gör. 


# Placera dina modulimpoter här: 
import matplotlib.pyplot as plt

import csv
# Deluppgift 1: Funktioner från deluppgift 1 i ordning.
# Skriv din kod här:
def read_file(filnamn): #Läser in csv-filen till en flerdimesionell lista, ändrar sedan datatyp på valda objekt från str till int. 
    
    with open(filnamn, 'r',encoding = 'UTF-8' ) as f:
        data_list = []
        csv_reader = csv.reader(f, delimiter = ';')
        for row in csv_reader:
            data_list.append(row)
        for row in data_list[2:]:
            n=1
            for i in row[1:]:
             row[n]= int(i)
             n+=1
     
    return data_list 
def first_five_rows(data):
    temp = []
    for row in data[:5]: 
        temp.append(row)
    print(temp)     
    return None
      
           

# Deluppgift 2: Funktioner från deluppgift 2 i ordning.
# Skriv din kod här:
def best_worst(data, index): #Går igenom CSV-datan och finner de 10 bästa samt sämsta länderna.
    kolum_list =[]
    land = ''
    resultat =0
    for row in data[2:]:        
        land =row[0]
        resultat = row[index]
        rad_list =[land,resultat]
        kolum_list.append(rad_list)
    toplist = sorted(kolum_list,key=lambda x:x[1], reverse=True)   #Sorterar listan efter resultat (fallande)
    top10 = []
    for row in toplist[:10]:
        top10.append(row) 
    bottomlist = sorted(kolum_list,key=lambda x:x[1])  #Sorterar listan efter resultat (Stigande)
    bott10= []
    for row in bottomlist[:10]:
        bott10.append(row)
    
    write_best_worst(top10, bott10) #Anropar utskriftsfunktion.


    return top10, bott10

def write_best_worst(tplist,btlist): # läser in listorna med de tio bästa/sämsta ländernas resultat. Skriver ut 'topplista' och 'sämstalista' i tabeller med tillhörande information.
    max_land = max(len(row[0]) for row in tplist)
    print(f'De tio länder som hade bäst resultat år 2018')
    print('---------------------------------------------')
    print('Land\t\tResultat')
    print('---------------------------------------------')
    for row in tplist:
        print(f'{row[0]}\t\t{str(row[1]).center(max_land)}')

    print('\n')

    print(f'De tio länder som hade sämst resultat år 2018')
    print('---------------------------------------------')
    print('Land\t\tResultat')
    print('---------------------------------------------')
    for row in btlist:
        print(f'{row[0]}\t\t{str(row[1]).center(10)}')       

# Deluppgift 3: Funktioner från deluppgift 3 i ordning.
# Skriv din kod här:


def kolumnmedel(data,index): #läser in csv-data och och index för valt år, räknar ut medelpoängen för vald kollumn i listan för csv-data.
    antal = 0
    sum = 0
    for row in data[2:]:
        
        
        varde = row[index]
        sum += varde
        antal +=1
    medel = sum/antal
    return medel



def arsmedel(data): #läser in csv-data, räknar ut årsmedelpoängen med hjälp av funktionen kolumnmedel. 
    armedel = []
    index =19
    for row in data[2:]:
        index -=1
        if index >=13:
            kolumnmedel(data,index)
            v = kolumnmedel(data,index)
            armedel.append(v)
    return armedel 
        
def country_year(v,w): #läser in csv-data och årsmedelvärdet,skapar listor för de nordiska länderna samt resultaten årsvis.
    Denmark = []
    Finland = []
    Iceland = []
    Norway  = []
    Sweden  = []
    for row in v[2:]:
        for i in row:
            if i == 'Denmark':
                Denmark.append(i)
                for j in range(13,19):
                    Denmark.append(row[j])
            elif i =='Finland':
                Finland.append(i)
                for j in range(13,19):
                    Finland.append(row[j])
            elif i =='Iceland':
                Iceland.append(i)
                for j in range(13,19):
                    Iceland.append(row[j])
            elif i == 'Norway':
                Norway.append(i)
                for j in range(13,19):
                    Norway.append(row[j])
            elif i == 'Sweden':
                Sweden.append(i)
                for j in range(13,19):
                    Sweden.append(row[j])

    list_2018 = ['2018',Sweden[1],Norway[1],Denmark[1],Finland[1],Iceland[1],f'{w[5]:.0f}']
    list_2015 = ['2015',Sweden[2],Norway[2],Denmark[2],Finland[2],Iceland[2],f'{w[4]:.0f}']
    list_2012 = ['2012',Sweden[3],Norway[3],Denmark[3],Finland[3],Iceland[3],f'{w[3]:.0f}']
    list_2009 = ['2009',Sweden[4],Norway[4],Denmark[4],Finland[4],Iceland[4],f'{w[2]:.0f}']
    list_2006 = ['2006',Sweden[5],Norway[5],Denmark[5],Finland[5],Iceland[5],f'{w[1]:.0f}']
    list_2003 = ['2003',Sweden[6],Norway[6],Denmark[6],Finland[6],Iceland[6],f'{w[0]:.0f}']
    return list_2018,list_2015,list_2012,list_2009,list_2006,list_2003, Denmark, Finland, Iceland, Norway, Sweden

def nordtabell(v,w): #läser in csv-data och årsmedelvärdet,skriver ut en tabell utifrån resultatet av de nordiska länderna från varje år.
    
    list_2018,list_2015,list_2012,list_2009,list_2006,list_2003, Denmark, Finland, Iceland, Norway, Sweden =country_year(v,w)
    
   
    print('Kunskapsutveckling i matematik enligt PISA-undersökningen 2003-2018.')
    print('--------------------------------------------------------------------')
    print('                             Länder')
    print('År\tSweden\tNorway\tDenmark\tFinland\tIceland\tMedelvärde\n\t\t\t\t\t\talla länder')
    print('--------------------------------------------------------------------')
    print(f'{list_2018[0]}\t{list_2018[1]}\t{list_2018[2]}\t{list_2018[3]}\t{list_2018[4]}\t{list_2018[5]}\t{list_2018[6]}')
    print(f'{list_2015[0]}\t{list_2015[1]}\t{list_2015[2]}\t{list_2015[3]}\t{list_2015[4]}\t{list_2015[5]}\t{list_2015[6]}')
    print(f'{list_2012[0]}\t{list_2012[1]}\t{list_2012[2]}\t{list_2012[3]}\t{list_2012[4]}\t{list_2012[5]}\t{list_2012[6]}')
    print(f'{list_2009[0]}\t{list_2009[1]}\t{list_2009[2]}\t{list_2009[3]}\t{list_2009[4]}\t{list_2009[5]}\t{list_2009[6]}')
    print(f'{list_2006[0]}\t{list_2006[1]}\t{list_2006[2]}\t{list_2006[3]}\t{list_2006[4]}\t{list_2006[5]}\t{list_2006[6]}')
    print(f'{list_2003[0]}\t{list_2003[1]}\t{list_2003[2]}\t{list_2003[3]}\t{list_2003[4]}\t{list_2003[5]}\t{list_2003[6]}')
    return list_2018,list_2015,list_2012,list_2009,list_2006,list_2003

def nordgraf(v,w): #skapar ett linjediagram för varje lands utveckling från 2003-2018.
    list_2018,list_2015,list_2012,list_2009,list_2006,list_2003, Denmark, Finland, Iceland, Norway, Sweden =country_year(v,w)
    år = [2003,2006,2009,2012,2015,2018]
    X_labels= (['2003', '2006', '2009', '2012', '2015', '2018'])
    plt.plot(år,Sweden [1:7],color = 'y', label = 'Sweden' )
    plt.plot(år,Denmark [1:7], color = 'r', label = 'Denmark' )
    plt.plot(år,Finland [1:7], color = 'b', label = 'Finland' )
    plt.plot(år,Norway [1:7], color = 'purple', label = 'Norway' )
    plt.plot(år,Iceland [1:7], color = 'g', label = 'Iceland' )
    plt.plot(år,w,color = 'black', label = 'Medelvärde' )
    plt.grid(color = 'black')
    plt.gca().set_facecolor('lightblue')
    plt.xlabel('År')
    plt.ylabel('Poäng')
    plt.title('PISA: Kunskapsutvecklingen i matematik 2003-2018')
    plt.xticks(år)

    plt.legend()
    plt.show()
    return None

# Deluppgift 4: Funktioner från deluppgift 4 i ordning.
# Skriv din kod här:

def battresammre(data,utveck): #undersöker utifall ett land i datamängden har en posetiv/negativ trend. Skriver sedan ut länder och resultat i tabeller.
    battre = []
    if utveck ==True:       
        i =0
        if i<6:
            for row in data[2:]:
                comp=1000
                for e in row[13:19]:
                    if e<=comp:
                        i+=1
                        comp= e
                        if i==6:
                            battre.append(row)
                            
                    else: 
                        i=0
                        break
        print('Länder som hela tiden har förbättras sina resultat mellan 2003-2018')
        print('--------------------------------------------------------------------')
        print('                             År och resultat:')
        print('Land\t\t\t2018\t2015\t2012\t2009\t2006\t2003')
        print('--------------------------------------------------------------------')
        print(f'{battre[0][0]}\t\t{battre[0][13]}\t{battre[0][14]}\t{battre[0][15]}\t{battre[0][16]}\t{battre[0][17]}\t{battre[0][18]}\t')                
       
                
                


    elif utveck== False:
        samre = []
        i =0
        if i<6:
            for row in data[2:]:
                comp=0
                for e in row[13:19]:
                    if e>=comp:
                        i+=1
                        comp= e
                        if i==6:
                            samre.append(row)
                            
                    else: 
                        i=0
                        break
        print('Länder som hela tiden har försämrat sina resultat mellan 2003-2018')
        print('--------------------------------------------------------------------')
        print('                             År och resultat:')
        print('Land\t\t\t2018\t2015\t2012\t2009\t2006\t2003')
        print('--------------------------------------------------------------------')
        print(f'{samre[0][0]}\t\t{samre[0][13]}\t{samre[0][14]}\t{samre[0][15]}\t{samre[0][16]}\t{samre[0][17]}\t{samre[0][18]}\t')
        print(f'{samre[1][0]}\t\t\t{samre[1][13]}\t{samre[1][14]}\t{samre[1][15]}\t{samre[1][16]}\t{samre[1][17]}\t{samre[1][18]}\t')                
        print(f'{samre[2][0]}\t\t{samre[2][13]}\t{samre[2][14]}\t{samre[2][15]}\t{samre[2][16]}\t{samre[2][17]}\t{samre[2][18]}\t')
    #print(battre)
    #print(samre)

    
    #print(f'{samre[13]}\t{samre[14]}\t{samre[15]}\t{samre[16]}\t{samre[17]}\t{samre[18]}\t')
    return None





# Deluppgift 5: Funktioner från deluppgift 5 i ordning.
# Skriv din kod här:


# Huvudprogram med Meny från deluppgift 0. Använd menyrubriker enl. uppgiftsbeskrivningen.
# Skriv din kod här:

state = True
while state:
    #Printar ut de olika menyvalen.
     print(f' \n\t1. Läs in CSV-filen. ')
     print(f' \t2. Bästa resp. sämsta resultatet år 2018.')
     print(f' \t3. Matematikkunskaper i norden år 2003 - 2018.')
     print(f' \t4. Kontinuerligt förbättrat resp. försämrat år 2003 – 2018.')
     print(f' \t5. Kvinnor presterar bättre än män under åren 2003–2018.')
     print(f' \t6. Avsluta programmet.')
     print(f'\n Välj ett menyalternativ (1-6) :')
     
     val = input()
     
     if val == '1':#Frågar användaren efter namnet på datafilen. Därefter läses csv-filen in och listan pisadata skapas enligt uppgift 1. Vi förutsätter att csv-filen ligger i arbetskatalogen för VC-code.
         filename =input('Ange filnamn eller tryck bara Enter för pisadata.csv') or 'pisadata.csv'
         pisadata = read_file(filename)
         first_five_rows (pisadata)
     
     elif val == '2':#Skapar tabeller över de tio läder med bästa resp. sämsta resultat 2018 enligt uppgift 2.
         index_2018 = [i for i, value in enumerate(pisadata[0]) if value == '2018'] #söker i datan efter index där 2018 och poängmedelvärdet sammanfaller.  

         # Hitta index för 'medel' i den andra raden i samma kolumn som '2018'
         index_medel = None
         for i, value in enumerate(pisadata[1]):
            if value == 'medel':
                index_medel = i
            break
         
         for idx in index_2018:
           if pisadata[1][idx] == 'medel':
               break
        
         pisadata_sorterad = best_worst(pisadata, idx)
    
     elif val == '3':#Skapar en tabell och ritar ett diagram över matematikkunskaperna i norden enligt uppgift 3.
        arsmedel(pisadata)
        medelvAr = (arsmedel(pisadata))
        print()
        nordtabell(pisadata,medelvAr)
        nordgraf(pisadata,medelvAr)
     elif val == '4':#Skapar tabeller över de länder där resultaten kontinuerligt förbättrats resp. försämrats enligt uppgift 4.
         T = True
         F = False
         battresammre(pisadata,T)
         print()
         battresammre(pisadata,F)
    
     elif val == '5': #Skapar en tabell på när kvinnor presterat bättre än män, enligt uppgift 5.
         print()
     elif val == '6': #Programmet avslutas och en bekräftelse t.ex. ”Programmet är avslutat” skrivs på skärmen.
         print('Programmet är avslutat')
         state = False
     
     else:
         print(f'Felaktig inmatning, försök igen!')
         
     if val !=6:
         nyvarv = input('\nVidare till menyn tryck Enter! ')
     
```
