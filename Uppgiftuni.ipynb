```python
# Skriv en inledande kommentar som talar om vad programmet gör.
#....
# Placera dina modulimpoter här:
import matplotlib
import csv

# Deluppgift 1: Funktioner från deluppgift 1 i ordning.
# Skriv din kod här:
data=[]
with open ("pisadata.csv", "r") as file: #Öppnar filen och sedan läser
    read = csv.reader(file, delimiter = ";")

    for rad in read: #läser filen och använder for sats för att lägga in i en lista
        data.append(rad)
 
def menytxt(): #Meny 
    print("1. Läs in csv-filen")
    print("2. Bästa resp. sämsta resultat år 2018")
    print("3. Matematikkunskaper i norden år 2003 – 2018.")
    print("4. Kontinuerligt förbättrat resp. försämrat år 2003 – 2018.")
    print("5. Kvinnor presterar bättre än män under åren 2003–2018")
    print("6. Avsluta programmet.")
    print("Välj ett menyalternativ (1-6):")
def meny(): # Använder whileloop för att skapa en meny. 
    while True:
        menytxt()
        val = int(input()) #frågar efter num och använder if satser "Meny" för att göra beslut
        if val == 1:
            filename=input("Ange filnamn eller tryck bara Enter för data.csv")          
            print(data)
            break
        elif val ==2:
            pisdatasorted()
            break
        elif val ==3:
            print(2)
            break
        elif val ==4:
            print(2)
            break
        elif val ==5:
            print(2)
            break       
        elif val ==6:
            break


# Deluppgift 2: Funktioner från deluppgift 2 i ordning.
# Skriv din kod här:
def år2018(l,i,TF): # Använder mig av val så det inte behvös repeteras 
    data.sort(reverse = TF, key= lambda i:i [1:]) # Sorterar datan från 2018+ använder sedan reverse för att fixa störst till minst
    for col in data[1:11]: # For sats för att skriva ut data från 1 till 10 
        year18 = col[0:3]
        year18.sort(reverse=True)# sorterar störst till minst mellan female och male mellan 2018
        print(f"{year18[l]}--{year18[i]}") #skriver ut med val om vilken lista som ska ut
    
def pisdatasorted():
    print("De tio länder som hade bäst resultat år 2018")
    print("----------------")
    print("Land   Resultat")
    print("----------------")
    år2018(0,1,True)
    print("----------------")
    print("De tio länder som hade sämst resultat år 2018")
    print("----------------")
    print("Land   Resultat")
    print("----------------")
    år2018(0,2,False)
    print("----------------")
#få medelvärde på alla länder
        


    
data.sort(reverse = False, key= lambda k:k[0]) #sort list


land = ["Sweden","Norway","Denmark","Finland","Iceland"] #skapar en lista för att söka på länder
länder = [] #en lista 

for w in land: # skapar for loops för att söka på de länder vi letar efter
    for rad in data:
        if rad[0] == w: #söker på rad 0 och sedan lägger till den i listan om det matchar listan
            länder.append(rad)
    for i in länder:
        del i[:-6]#tar bort allt förutom medelvärden från norden
      

medelländer= [] #skapar lista
for x in data:#använder mig av for loops för att filtrerar ut
    del x[:-5] 
    medelländer.append(x[0:])
år = medelländer[0]# Skapar värden för 2018 till 2003
del medelländer[0:2]# tar bort land och M/K

og=[]
for k in medelländer:
    og.append(k)

og_fix = list(zip(*og)) #Vänder på listan så att det går och räkna från 2003 till 2018
all_rows = []
row_sums = []
for column in og_fix: #skapar foor loops i varandra och för att lägga som i var sin lista 
    row_list = []
    row_sum = 0  
    for value in column:
        row_list.append(int(value))
        row_sum += int(value)#skapar en sum funktion för varje lista 
    all_rows.append(row_list)
    row_sums.append(row_sum)

medelvärded= []
for row, row_sum in zip(all_rows, row_sums): #beräknar medel värdet för 2003 till 2018 och sedan lägger det i en lisa med hjälp av for loops
    medelvärde = round(row_sum/len(row))#använder mig av len för att att beräkna antal värden per lista
    medelvärded.append(medelvärde)



print("\nKunskapsutveckling i matematik enligt PISA-undersökningen 2003 – 2018.")
print("                               Länder:")
print("--------------------------------------------------------------------------------------")
print("År  Sweden Norway Denmark Finland Iceland          Medelvärde alla länder")
print("--------------------------------------------------------------------------------------")

for år, länder,medelvärded in zip(år,länder,medelvärded):
    print(f"{år}  {länder}                 {medelvärded} ")

#print(allaländer)
# Deluppgift 3: Funktioner från deluppgift 3 i ordning.
# Skriv din kod här:
# Deluppgift 4: Funktioner från deluppgift 4 i ordning.
# Skriv din kod här:
# Deluppgift 5: Funktioner från deluppgift 5 i ordning.
# Skriv din kod här:
# Huvudprogram med Meny från deluppgift 0. Använd menyrubriker enl.uppgiftsbeskrivningen.
# Skriv din kod här:
```
